```
import sklearn.metrics as metrics
import matplotlib.pyplot as plt
```

```

def plot_roc_curve(fpr, tpr, model_name = 'CNN', name='AUC', color='b'):
  roc_auc = metrics.auc(fpr, tpr)
  import matplotlib.pyplot as plt
  plt.title('ROC Curve for ' + model_name)
  plt.plot(fpr, tpr, color, label = name + ' = %0.2f' % roc_auc)
  plt.legend(loc = 'lower right')
  plt.plot([0, 1], [0, 1],'r--')
  plt.xlim([0, 1])
  plt.ylim([0, 1])
  plt.ylabel('True Positive Rate')
  plt.xlabel('False Positive Rate')
```

```
cnn_pre_fpr, cnn_pre_tpr, cnn_pre_threshold = [0., 0.01044634, 1.], [0., 0.4303304, 1.], [2, 1, 0]
cnn_cust_fpr, cnn_cust_tpr, cnn_cust_threshold = [0., 0.01593882, 1.], [0., 0.45402797, 1.], [2, 1, 0]
cnn_rand_fpr, cnn_rand_tpr, cnn_rand_threshold = [0., 0.01296815, 1.], [0., 0.47259003, 1.], [2, 1, 0]

lstm_pre_fpr, lstm_pre_tpr, lstm_pre_threshold = [0., 0.01657132, 1.], [0., 0.61941591, 1.], [2, 1, 0]
lstm_cust_fpr, lstm_cust_tpr, lstm_cust_threshold = [0., 0.01982764, 1.], [0., 0.50761044, 1.], [2, 1, 0]
lstm_rand_fpr, lstm_rand_tpr, lstm_rand_threshold = [0., 0.0081612, 1.], [0., 0.17974261, 1.], [2, 1, 0]

gru_pre_fpr, gru_pre_tpr, gru_pre_threshold = [0., 0.01657132, 1.], [0., 0.61941591, 1.], [2, 1, 0]
gru_cust_fpr, gru_cust_tpr, gru_cust_threshold = [0., 0.01484114, 1.], [0., 0.44988244, 1.],[2, 1, 0]
gru_rand_fpr, gru_rand_tpr, gru_rand_threshold = [0., 0.00605967, 1.], [0., 0.15326074, 1.], [2, 1, 0]



```

```
plot_roc_curve(cnn_pre_fpr, cnn_pre_tpr, "CNN", "Pre Trained", 'b')
plot_roc_curve(cnn_cust_fpr, cnn_cust_tpr, "CNN",  "Custom", 'g')
plot_roc_curve(cnn_rand_fpr, cnn_rand_tpr, "CNN", "Random", 'y')

plt.show()
```

```
plot_roc_curve(lstm_pre_fpr, lstm_pre_tpr, "LSTM", "Pre Trained", 'b')
plot_roc_curve(lstm_cust_fpr, lstm_cust_tpr, "LSTM", "Custom", 'g')
plot_roc_curve(lstm_rand_fpr, lstm_rand_tpr, "LSTM", "Random", 'y')

plt.show()
```
